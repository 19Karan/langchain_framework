```python
from langchain.chat_models import init_chat_model
model=init_chat_model("groq:meta-llama/llama-4-scout-17b-16e-instruct")
```

```python
from pydantic import BaseModel, Field
class Movie(BaseModel):
    title: str = Field(description="The title of the movie")
    director: str = Field(description="The director of the movie")
    release_year: int = Field(description="The year the movie was released")
    rating: float = Field(description="The rating of the movie")   
```

```python
model_struct=model.with_structured_output(Movie)
model_struct.invoke("provide details about the interstellar movie")
```

```python
from pydantic import BaseModel, Field
class Movie(BaseModel):
    title: str = Field(..., description="The title of the movie")
    director: str = Field(..., description="The director of the movie")
    release_year: int = Field(..., description="The year the movie was released")
    rating: float = Field(..., description="The rating of the movie")   
```

```python
model_struct1=model.with_structured_output(Movie,include_raw=True)
model_struct1.invoke("provide details about the titanic movie")
```
