```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models  import init_chat_model
model=init_chat_model("groq:meta-llama/llama-4-scout-17b-16e-instruct")
```

```python
model.invoke("who is the apple")
```

# Streaming


```python
model.invoke("What is ai explain 200 words?")
```

```python
for chunk in model.stream("What is ai explain 200 words?"):
    print(chunk.text,end="",flush=True)
```

```python
for chunk in model.stream("why parrot have colorful features?"):
    print(chunk.text,end="",flush=True)
```

```python
response=model.invoke("why parrot have colorful features?")
```

```python
print(response.text)
```

```python
# Batch
responses=model.batch(["why parrot have colorful features?","What is ai explain 200 words?"])
```

```python
for response in responses:
    print(response.text)
```

```python
model.batch(
    ["who is the currant pm of india?","who is the currant president of india?","who is the fonder of chatGPT?"],
    config={"max_concurrency":5} #Limit to 5 parallel calls
    )
```
